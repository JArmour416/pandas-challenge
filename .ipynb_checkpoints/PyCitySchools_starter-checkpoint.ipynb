```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data_df = pd.read_csv(school_data_to_load, encoding="iso-8859-1")
student_data_df = pd.read_csv(student_data_to_load, encoding="iso-8859-1")
school_data_df.head()
```

```python
student_data_df.head()
```

```python
# Combine the data into a single dataset
school_data_complete_df = pd.merge(school_data_df, student_data_df, on=["school_name"], how="left")
school_data_complete_df.head()
```

```python
#### District Summary

# Find the total schools, students and budget
school_total = school_data_df["school_name"].nunique()
student_total = student_data_df["student_name"].count()
budget_total = school_data_df["budget"].sum()
                
# Find the average math and reading scores
math_score_avg = student_data_df["math_score"].mean()
reading_score_avg = student_data_df["reading_score"].mean()

# Set Boolean variables to find passing (true/false)
student_data_df["Math Pass"] = student_data_df["math_score"]> 69
student_data_df["Reading Pass"] = student_data_df["reading_score"]> 69

# converting dtypes using astype 1 for pass and zero fail
student_data_df["Math Pass"]= student_data_df["Math Pass"].astype(int) 
student_data_df["Reading Pass"]= student_data_df["Reading Pass"].astype(int)  

student_data_df
```

```python
# Find percentages of students passing math, passing reading and overall passing
math_pass = student_data_df["Math Pass"].sum()
math_pass_percent = ((math_pass/student_total)*100)
reading_pass=student_data_df["Reading Pass"].sum()
reading_pass_percent = ((reading_pass/student_total)*100)
overall_pass = ((math_score_avg + reading_score_avg)/2)
 
# Create a dataframe
district_summary_df = pd.DataFrame({"Total Schools": [school_total],
                                    "Total Students": [student_total],
                                    "Total Budget": [budget_total],
                                    "Average Math Score": [math_score_avg],
                                    "Average Reading Score": [reading_score_avg],
                                    "% Passing Math": [math_pass_percent],
                                    "% Passing Reading": [reading_pass_percent],
                                    "Overall Passing Rate": [overall_pass]})
district_summary_df = district_summary_df[["Total Schools", "Total Students", "Total Budget",
                                    "Average Math Score", "Average Reading Score", "% Passing Math",
                                    "% Passing Reading", "Overall Passing Rate"]]
district_summary_df=district_summary_df.round(2)
district_summary_df["Total Students"]=district_summary_df["Total Students"].map("{0:,d}".format)
district_summary_df["Average Math Score"]=district_summary_df["Average Math Score"].map("{0:,.2f}".format)
district_summary_df["Average Reading Score"]=district_summary_df["Average Reading Score"].map("{0:,.2f}".format)
district_summary_df["% Passing Math"]=district_summary_df["% Passing Math"].map("{0:,.2f}".format)
district_summary_df["% Passing Reading"]=district_summary_df["% Passing Reading"].map("{0:,.2f}".format)
district_summary_df["Overall Passing Rate"]=district_summary_df["Overall Passing Rate"].map("{0:,.2f}".format)
district_summary_df["Total Budget"]=district_summary_df["Total Budget"].map("${0:,.0f}".format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#### School Summary

# Create a dataframe for school summary
school_summary_df = school_data_df

# Rename headers
school_summary_df = school_summary_df.rename(columns={'type':'School Type', 'size':'Total Students', 
                                                  'budget':'Total School Budget', 'math_score':'Math Score',
                                                 'reading_score':'Reading Score'})
# Delete school ID 
del school_summary_df['School ID']
```

```python
# Create per student budget and move it forward
school_summary_df['Per Student Budget'] = school_summary_df['Total School Budget']/school_summary_df['Total Students']

# Groupby schools the student data
school_students_gp = student_data_df.groupby(['school_name'])
school_summary_df.head()
```

```python
# Set variables in order to create overall pass rate, % passing math and reading
school_math_pass = school_students_gp['math_score'].mean()
school_reading_pass = school_students_gp['reading_score'].mean()
school_math_total = school_students_gp['Math Pass'].sum()
school_reading_total = school_students_gp['Reading Pass'].sum()
school_math_df = pd.DataFrame(school_math_pass)
school_math_df['Math Pass Total'] = school_math_total
school_reading_df = pd.DataFrame(school_reading_pass)
school_reading_df['Reading Pass Total'] = school_reading_total

# merge dataframes and create overall pass rate, % passing math and reading
school_both_df = pd.merge(school_math_df, school_reading_df, on='school_name') 
school_summary_df = pd.merge(school_summary_df, school_both_df, on='school_name', how='inner')
school_summary_df['% Passing Math'] = (school_summary_df['Reading Pass Total'])/school_summary_df['Total Students']*100
school_summary_df['% Passing Reading'] = (school_summary_df['Math Pass Total']) /(school_summary_df['Total Students'])*100
school_summary_df['Overall Passing Rate'] = ((school_summary_df['% Passing Math']+ school_summary_df['% Passing Math'])/2)

# Format the result to read more easily
school_summary_df = school_summary_df.rename(columns={'school_name':'School Name', 'math_score':'Average Math Score', 
                                                        'reading_score':'Average Reading Score'})

# Delete math and reading pass totals
del school_summary_df['Math Pass Total']
del school_summary_df['Reading Pass Total']

# More formatting
school_summary_df = school_summary_df.round(2)
school_summary_df["Total Students"]=school_summary_df["Total Students"].map("{0:,d}".format)
school_summary_df["Total School Budget"]=school_summary_df["Total School Budget"].map("${0:,.0f}".format)
school_summary_df["Per Student Budget"]=school_summary_df["Per Student Budget"].map("${0:,.0f}".format)
school_summary_df["Average Math Score"]=school_summary_df["Average Math Score"].map("{0:,.2f}".format)
school_summary_df["Average Reading Score"]=school_summary_df["Average Reading Score"].map("{0:,.2f}".format)
school_summary_df["% Passing Math"]=school_summary_df["% Passing Math"].map("{0:,.2f}".format)
school_summary_df["% Passing Reading"]=school_summary_df["% Passing Reading"].map("{0:,.2f}".format)
school_summary_df["Overall Passing Rate"]=school_summary_df["Overall Passing Rate"].map("{0:,.2f}".format)
school_summary_df

school_summary_df.head()
```
